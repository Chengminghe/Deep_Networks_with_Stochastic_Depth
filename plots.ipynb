```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
##training history of cifar10
with open('./Logs/ResNet110_cifar10.npy', 'rb') as f:
     log1 = np.load(f)
with open('./Logs/StochasticNet110_cifar10.npy', 'rb') as f:
     log2 = np.load(f)
```

```python
plt.plot(1-log1[1],'r--',label='training error')
plt.plot(1-log1[2],label='validation error',color='r')
plt.plot(1-log2[1],'g--',label='training error(stochastic)')
plt.plot(1-log2[2],label='validation error(stochastic)',color='g')
plt.ylim(0,0.4)
plt.xlabel('epochs')
plt.ylabel('error')
plt.legend()
total_time_1 = 24874.805114269257
total_time_2 = 21912.673411130905
```
